## Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

Ans) In Object-Oriented Programming (OOP), a class is a blueprint or template that defines the attributes and behaviors of a set of objects. An object is an instance of a class that contains data and methods.

A class defines the properties (attributes) and actions (methods) that objects of that class will have. For example, a class "Person" can have attributes like name, age, gender, and methods like walk, talk, and eat.

An object is an instance of a class, created using the constructor of the class. For example, we can create an object of the "Person" class with a specific name, age, and gender.

Here's an example of a "Person" class in Python:

```python
class person:
    def __init__(self, name, age, gender):
        self.name= name
        self.age= age
        self.gender= gender
        
    def get_person_details(self):
        return self.name, self.age, self.gender
    
    def walk(self):
        print(self.name + " walks in a high speed she's " + self.gender)  
```

```python
a=person("ananya",56,"Female")
```

```python
a.name
```

```python
a.age
```

```python
a.gender
```

```python
a.get_person_details()
```

```python
a.walk()
```

## Q2. Name the four pillars of OOPs. 

Ans) The four pillars of Object-Oriented Programming (OOP) are:

<b><u><font color="green">Encapsulation:</font> </u></b> 
Encapsulation is the technique of hiding the internal details of an object and exposing only the necessary information to the outside world. It helps to protect the data from being accidentally modified, and allows us to change the internal implementation of an object without affecting other parts of the code.

<b><u><font color="green">Abstraction:</font></u></b>
Abstraction is the process of identifying and extracting essential features of an object and ignoring the rest. It allows us to create abstract classes and interfaces that define a set of methods without implementing them, and leave the implementation details to the subclasses.

<b><u><font color="green">Inheritance:</font></u></b>
Inheritance is the mechanism of creating a new class by inheriting the properties (attributes and methods) of an existing class. It allows us to reuse code and create a hierarchy of classes with increasing levels of specialization.

<b><u><font color="green">Polymorphism:</font></u></b>
Polymorphism is the ability of objects of different classes to be used interchangeably. It allows us to write code that can work with objects of multiple classes without knowing their exact types, and provides flexibility and extensibility to the code.

## Q3. Explain why the __init__() function is used. Give a suitable example.

Ans) In object-oriented programming, the __init__() function is used as a constructor method for a class. It is called automatically when an instance of the class is created and is used to initialize the attributes of the object with default values or values provided by the user.

Here's an example to illustrate why the __init__() function is used:

```python
class person:
    def __init__(shru, name, age):
        shru.name= name
        shru.age= age
        shru.gender= "unknown"
```

```python
s=person("shruti",25)
```

```python
s.name
```

```python
s.age
```

```python
s.gender
```

In this example, we create a Person object named s with the name "Shruti" and age 25. Since we did not provide a value for the gender attribute, it is initialized to the default value of "Unknown". We can access the values of the attributes using dot notation.

The __init__() function is useful for setting up the initial state of an object, especially when the object has multiple attributes that need to be initialized with specific values. By defining an __init__() method for a class, we can ensure that every object created from that class is initialized correctly and consistently.

## Q5. What is inheritance? Give an example for each type of inheritance.

Ans) Inheritance is a concept in Object-Oriented Programming that allows a new class to be based on an existing class, inheriting its attributes and methods. The existing class is called the parent or base class, and the new class is called the child or derived class. The child class can add new attributes and methods, or modify the ones inherited from the parent class.

There are several types of inheritance in OOP. Here are some examples:

<u>Single inheritance:</u>
Single inheritance is when a child class inherits from a single parent class. This is the most common type of inheritance. For example:

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} makes a sound.")
    
class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks.")
    
dog1 = Dog("Rufus")
dog1.speak() # output: Rufus barks.
```

In this example, we define an Animal class with an __init__() method and a speak() method. We then define a Dog class that inherits from the Animal class and overrides its speak() method to make the dog bark instead of just making a sound. We create a Dog object named dog1 with the name "Rufus" and call its speak() method, which prints out the message "Rufus barks.".

<u>2)Multi-level inheritance:</u>
Multi-level inheritance is when a child class inherits from a parent class, which in turn inherits from another parent class. For example:

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} makes a sound.")
    
class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks.")
    
class Labrador(Dog):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color
    
    def fetch(self):
        print(f"{self.name} fetches the ball.")
    
labrador1 = Labrador("Max", "yellow")
labrador1.speak() # output: Max barks.
labrador1.fetch() # output: Max fetches the ball.
```

In this example, we define an Animal class and a Dog class as in the previous example. We then define a Labrador class that inherits from the Dog class and adds a new fetch() method. We create a Labrador object named labrador1 with the name "Max" and color "yellow". We call its speak() method, which prints out the message "Max barks.", and its fetch() method, which prints out the message "Max fetches the ball.".

3) <u>Multiple inheritance:</u> Multiple inheritance is when a child class inherits from multiple parent classes. For example:

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} makes a sound.")
    
class Pet:
    def __init__(self, owner):
        self.owner = owner
    
    def feed(self):
        print(f"{self.owner} feeds {self.name}.")
    
class Dog:
    def speak(self, Animal, Pet):
        print(f"{self.name} barks.")
    
```
